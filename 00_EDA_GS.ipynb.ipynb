## 파일 병합
- a_ms와 a_ind를 먼저 병합하고 그 뒤에 a_hh를 병합해서 a_full 생성

```python
# a_ms와 a_ind간 merge를 'HHID', 'PIDWON'으로 병합
a_first = pd.merge(a_ms, a_ind, on=["HHID", "PIDWON"], how="outer")
```

```python
# 위에 병합했던 a_first를 a_hh랑 merge해서 'HHID'로 병합
a_full = pd.merge(a_first, a_hh, on="HHID", how="outer")
```

## 나이컬럼 생성
- 태어난 년도와 태어난 월을 현재 나이 기준으로 만나이 적용해서 새로운 나이컬럼 AGE 생성

```python
# 태어난 년도와 월을 기준으로 나이 정리
from datetime import datetime   

def calculate_age(row):
    today = datetime.today()
    birth_year = row['BIRTH_Y']
    birth_month = row['BIRTH_M']
    if pd.isnull(birth_year) or pd.isnull(birth_month):
        return None
    age = today.year - birth_year - ((today.month, today.day) < (birth_month, 1))
    return age

a_full['AGE'] = a_full.apply(calculate_age, axis=1)
a_full['AGE'].to_frame()
```

```python
# 파생변수 AGE를 만들어서 필요없어진 BIRTH_Y, BIRTH_M은 제거
a_full.drop(['BIRTH_Y','BIRTH_M'], axis=1, inplace=True, errors='ignore')
```

## 컬럼이름 변경

## 의료서비스 데이터 만들기

```python
# . 의료서비스 데이터셋 만들기
#    - mexp3_1 값이 -9인 경우 0으로 치환 (모름/무응답 처리)
a_full["MEXP3_1"] = a_full["MEXP3_1"].where(a_full["MEXP3_1"] != -9, 0)
```

## 집계함수 적용
- 데이터를 PIDWON(개인 식별자)와 장애유무(장애 여부) 컬럼을 기준으로 그룹화
- 그후 집계함수 적용

```python
#  개인별( pidwon ) 건수, 총이용일수, 총의료비 집계 (파생변수 생성)
#    - 건수: 동일 pidwon의 행 개수로 계산
#    - 총이용일수: days의 합계
#    - 총의료비: mexp3_1의 합계
#    - DISA_TY: 그룹 내 첫 번째 행에 있는 장애유형 값을 그대로 가져온다

agg_a_full = a_full.groupby(["PIDWON", "장애유무"], as_index=False).agg(
    건수=("PIDWON", "count"),
    총이용일수=("DAYS", "sum"),
    총의료비=("MEXP3_1", "sum"),
    DISA_TY=("장애유형", "first"),
    나이=("나이", "first"),
    성별=("성별", "first")
)
```

## a_full_info 생성 및 병합

```python
#  분석용 데이터셋(ind_ms) 만들기
#    - 이미 a_full에는 pidwon, region1, sex 등의 정보가 있으므로,
#      중복되지 않도록 pidwon기준으로 병합
# -------------------------------------------
a_full_info = a_full.drop_duplicates(subset="PIDWON", keep="first")[["PIDWON", "지역", "성별"]]
ind_ms = pd.merge(a_full_info, agg_a_full, on="PIDWON", how="outer")
```

## use 변수 생성

```python
#  결측값 처리 및 'use' 변수 생성
#    - 건수, 총이용일수, 총의료비 결측값은 0으로 대체
#    - 건수 >= 1 이면 use=1 (의료이용자), 아니면 use=2
ind_ms["건수"] = ind_ms["건수"].fillna(0)
ind_ms["총이용일수"] = ind_ms["총이용일수"].fillna(0)
ind_ms["총의료비"] = ind_ms["총의료비"].fillna(0)
ind_ms["use"] = ind_ms["건수"].apply(lambda x: 1 if x >= 1 else 2)
```

## 나이컬럼 생성
- 태어난 년도와 태어난 월을 현재 나이 기준으로 만나이 적용해서 새로운 나이컬럼 AGE 생성

```python
# 태어난 년도와 월을 기준으로 나이 정리
from datetime import datetime   

def calculate_age(row):
    today = datetime.today()
    birth_year = row['BIRTH_Y']
    birth_month = row['BIRTH_M']
    if pd.isnull(birth_year) or pd.isnull(birth_month):
        return None
    age = today.year - birth_year - ((today.month, today.day) < (birth_month, 1))
    return age

b_full['AGE'] = b_full.apply(calculate_age, axis=1)
b_full['AGE'].to_frame()
```

## 집계함수 적용
- 데이터를 PIDWON(개인 식별자)와 장애유무(장애 여부) 컬럼을 기준으로 그룹화
- 그후 집계함수 적용

```python
#  개인별( pidwon ) 건수, 총이용일수, 총의료비 집계 (파생변수 생성)
#    - 건수: 동일 pidwon의 행 개수로 계산
#    - 총이용일수: days의 합계
#    - 총의료비: mexp3_1의 합계
#    - DISA_TY: 그룹 내 첫 번째 행에 있는 장애유형 값을 그대로 가져온다

agg_b_full = b_full.groupby(["PIDWON", "장애유무"], as_index=False).agg(
    건수=("PIDWON", "count"),
    총이용일수=("DAYS", "sum"),
    총의료비=("MEXP3_1", "sum"),
    DISA_TY=("장애유형", "first"),
    나이=("나이", "first"),
    성별=("성별", "first")
)
```

## 나이컬럼 생성
- 태어난 년도와 태어난 월을 현재 나이 기준으로 만나이 적용해서 새로운 나이컬럼 AGE 생성

```python
# 태어난 년도와 월을 기준으로 나이 정리
from datetime import datetime   

def calculate_age(row):
    today = datetime.today()
    birth_year = row['BIRTH_Y']
    birth_month = row['BIRTH_M']
    if pd.isnull(birth_year) or pd.isnull(birth_month):
        return None
    age = today.year - birth_year - ((today.month, today.day) < (birth_month, 1))
    return age

c_full['AGE'] = c_full.apply(calculate_age, axis=1)
c_full['AGE'].to_frame()
```

## 집계함수 적용
- 데이터를 PIDWON(개인 식별자)와 장애유무(장애 여부) 컬럼을 기준으로 그룹화
- 그후 집계함수 적용

```python
#  개인별( pidwon ) 건수, 총이용일수, 총의료비 집계 (파생변수 생성)
#    - 건수: 동일 pidwon의 행 개수로 계산
#    - 총이용일수: days의 합계
#    - 총의료비: mexp3_1의 합계
#    - DISA_TY: 그룹 내 첫 번째 행에 있는 장애유형 값을 그대로 가져온다

agg_c_full = c_full.groupby(["PIDWON", "장애유무"], as_index=False).agg(
    건수=("PIDWON", "count"),
    총이용일수=("DAYS", "sum"),
    총의료비=("MEXP3_1", "sum"),
    DISA_TY=("장애유형", "first"),
    나이=("나이", "first"),
    성별=("성별", "first")
)
```

## 장애 유형별 총 의료비 분포 시각화

```python
# 장애 유형 코드 
disa_map = {
    1.0: "지체",
    2.0: "뇌병변",
    3.0: "시각",
    4.0: "청각",
    5.0: "언어",
    6.0: "안면",
    7.0: "신장",
    8.0: "심장",
    9.0: "간",
    10.0: "호흡기",
    11.0: "장루·요루",
    12.0: "뇌전증",
    13.0: "발달(자폐성X)",
    14.0: "발달(자폐성)",
    15.0: "정신장애"
}

# 2. agg_c_full에 장애 유형 이름 컬럼 생성
agg_c_full["장애유형이름"] = agg_c_full["DISA_TY"].map(disa_map)

# 3. 박스플롯 시각화
plt.figure(figsize=(8,6))
ax = sns.boxplot(data=agg_c_full, x="장애유형이름", y="총의료비", palette="Set2")

plt.title("장애 유형별 총 의료비 분포")
plt.xlabel("장애 유형")
plt.ylabel("총 의료비")

# 4. y축 눈금에 천단위 콤마 적용
yvals = ax.get_yticks()
ax.set_yticklabels([f"{val:,.0f}" for val in yvals])

plt.xticks(rotation=45)   # x축 레이블 기울이기
plt.tight_layout()

# 5. 각 장애 유형별 총의료비 평균 계산 후, 박스플롯 위에 평균선 추가
group_means = agg_c_full.groupby("장애유형이름")["총의료비"].mean()

# 각 x축 카테고리 위치는 0부터 시작하는 정수라고 가정합니다.
for i, label in enumerate(ax.get_xticklabels()):
    cat = label.get_text()           # 해당 장애 유형 이름
    mean_val = group_means.get(cat)    # 해당 장애 유형의 총의료비 평균
    if mean_val is not None:
        # x축 위치 i의 중앙을 기준으로 ±0.3 범위에서 수평 점선을 그림
        # 첫번째 범주에 대해서만 범례를 추가하여 한 번만 표시되도록 합니다.
        if i == 0:
            plt.plot([i-0.3, i+0.3], [mean_val, mean_val],
                     color='red', linestyle='--', linewidth=2, label='평균')
        else:
            plt.plot([i-0.3, i+0.3], [mean_val, mean_val],
                     color='red', linestyle='--', linewidth=2)

plt.legend()
plt.show()
```
